```python
from pyspark.sql.types import StructType,StructField,StringType,IntegerType,DateType,FloatType,TimestampType,DoubleType

custom_schema=StructType([StructField("productId",StringType(),False),StructField("productName",StringType(),True),StructField("stdCost",FloatType(),True),StructField("stdPrice",DoubleType(),True),
                          StructField("effDt",TimestampType(),True),StructField("dt",DateType(),True)])
dfd = spark.read.schema(custom_schema).csv("/Volumes/telecom_catalog_assign/landing_zone/landing_vol/sales.csv",header=True)
display(dfd)
```
